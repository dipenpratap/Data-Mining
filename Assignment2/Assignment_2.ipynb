 # Problem #1
 1. Using the churn data set, determine how many records need to be resampled in order to have 20% of the rebalanced dataset have true Churn variable values. Create the rebalanced data set and confirm that the new data set has 20% true Churn variable values.


```python
# Balancing the data to get 20% of the churn variables to have true values
n = dataset.shape[0]
Churn = dataset['Churn']
false,true = Churn.value_counts()
```

```python
# Calculating the number of records to be resampled
num_to_add = int((0.2*n-true)/(1-0.2))
print('The number of records to be resampled is {}.'.format(num_to_add))
```

```python
to_resample_loc = dataset.loc[dataset['Churn']==True]
to_resample_loc
```

```python
# Creating the rebalanced dataset
new_samples = to_resample_loc.sample(n=num_to_add,replace = True,random_state = 7)
print('The new samples for rebalance are: ')
new_samples
```

```python
# Concatenating the samples with original dataset
dataset_rebal = pd.concat([dataset,new_samples])
print('The new rebalanced dataset is: ')
dataset_rebal
```

```python
new_n = dataset_rebal.shape[0]
new_false,new_true = dataset_rebal['Churn'].value_counts()
new_true_per = np.round((new_true/new_n)*100,decimals=1)
print('The new dataset has {} percent {} Churn instances.'.format(new_true_per,'True'))
```

### Second part of the problem 
"""Validate the training and test data sets by testing for the difference
in the training and test means using Day.Mins (t-test) and the Z-test on the Churn variable. Try forming new training and test
sets if there is enough evidence to reject the null hypothesis.
"""

```python
# finding the mean and standard deviation of training data using Day_mins
day_mins_train = train_data['Day Mins']
mean_train_day_mins = st.mean(day_mins_train)
sd_train_day_mins =st.stdev(day_mins_train)


# finding the mean and standard deviation of test data using Day Mins
day_mins_test = test_data['Day Mins']
mean_test_day_mins = st.mean(day_mins_test)
sd_test_day_mins = st.stdev(day_mins_test)

# Number of isntances in the train and test data
train_n = train_data.shape[0]
test_n = test_data.shape[0]


s1_n1 = (sd_train_day_mins**2)/train_n
s2_n2 = (sd_test_day_mins**2)/test_n



t_Test_values = pd.DataFrame({'Data Type':['Train','Test'],'n':[train_n,test_n],
                             'mean':[mean_train_day_mins,mean_test_day_mins],
                             'Standard Deviation':[sd_train_day_mins,sd_test_day_mins]})

print('The values required for the calculation of t_Test value are: ')
t_Test_values
```

```python
t_Test_Val = (mean_train_day_mins - mean_test_day_mins)/np.sqrt(s1_n1+s2_n2)
print('The t_Test value is {}'.format(t_Test_Val))
```

```python
# Alternatively, the p value and t-test value calculated using inbuilt function
t_val2,p_val_2= stats.ttest_ind(day_mins_train,day_mins_test)
print('The t_Test value is {}\n and the p_value is {}'.format(t_val2,p_val_2))
```

```python
# Z-Test on the Churn variable

"""
For a flag variable (like no/yes) we need the two-sample Z test for the
difference in proportions
Therefore we are going to perform a Z-test on the Churn variable as it is a flag variable that indicates either True 
or False"""
```

```python
train_churn = train_data['Churn']
test_churn = test_data['Churn']

# Getting the counts of true or false
false_train,true_train = train_churn.value_counts()
false_test,true_test = test_churn.value_counts()

# Lets proceed with true data
p_1 = true_train/train_n
p_2 = true_test/test_n
p_pooled = (true_train+true_test)/(train_n+test_n)
A = (1/train_n)+(1/test_n)
z_score = (p_1-p_2)/(np.sqrt(p_pooled*(1-p_pooled)*A))
```

```python
print('The z_data value is {}'.format(z_score))


# finding the p_value
p_values_z = stats.norm.sf(abs(z_score))*2
print('The p-value of the z-score is {}'.format(p_values_z))
```

# Problem #3
Create a CART model using the training set with the Churn target variable and whatever predictor variables you think
appropriate. Try at least 3 different models. Compare the confusion tables and accuracies of the 3 different models.

```python
import statsmodels.tools.tools as stattools
from sklearn.tree import DecisionTreeClassifier,export_graphviz
```

### Model 1

```python
# For simplicity lets save the Churn variable as target
target = train_data['Churn']

# We have a categorical variable, Intl Plan, among our predictors. The
# CART model implemented in the sklearn package needs categorical variables converted
# to a dummy variable form. Thus, we will make a series of dummy variables
# for Intl Plan using the categorical() command.

# Model 1

intl_plan_np = np.array(train_data['Intl Plan'])
intl_plan_cat, intl_plan_cat_dict = stattools.categorical(intl_plan_np,drop=True, dictnames = True)

day_charge = train_data[['Day Charge']]

intl_plan_cat_pd = pd.DataFrame(intl_plan_cat,index = day_charge.index)
X = pd.concat((train_data[['Day Charge']], intl_plan_cat_pd), axis = 1)


X_names = ["Day Charges", "Int Plan NO ", "Intl Plan YES"]
target_names = ["Churn False", "Churn True"]

# Now, we are ready to run the CART algorithm!

cart01 = DecisionTreeClassifier(criterion = "gini", max_leaf_nodes=5).fit(X,target)

export_graphviz(cart01, out_file = "cart01.dot",
feature_names=X_names, class_names=target_names)




#predIncomeCART is a numpy array so make it a dataframe
target = pd.DataFrame(target)
pred_int_plan_CART = cart01.predict(X)
pred_int_plan_CART_pd = pd.DataFrame(pred_int_plan_CART,index = day_charge.index)
pred_CART_target = pd.concat((target,pred_int_plan_CART_pd), axis = 1)


```

```python
import graphviz

with open("cart01.dot") as f:
    dot_graph = f.read()

graphviz.Source(dot_graph)
```

```python
#create a mask showing where the values differ
mask = target.where(target.values==pred_int_plan_CART_pd.values).notna()
model_predict = mask['Churn'].value_counts()
orig_data = train_data['Churn'].value_counts()
print('Masked values:')
model_predict
```
