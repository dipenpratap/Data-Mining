```python
# Importing the libraries
import pandas as pd
import numpy as np
import statistics as st
```

 1. Using the churn data set, determine how many records need to be resampled in order to have 20% of the rebalanced dataset have true Churn variable values. Create the rebalanced data set and confirm that the new data set has 20% true Churn variable values.


```python
# Balancing the data to get 20% of the churn variables to have true values
n = dataset.shape[0]
Churn = dataset['Churn']
false,true = Churn.value_counts()
```

```python
# Calculating the number of records to be resampled
num_to_add = int((0.2*n-true)/(1-0.2))
print('The number of records to be resampled is {}.'.format(num_to_add))
```

```python
to_resample_loc = dataset.loc[dataset['Churn']==True]
to_resample_loc
```

```python
# Creating the rebalanced dataset
new_samples = to_resample_loc.sample(n=num_to_add,replace = True)
print('The new samples for rebalance are: ')
new_samples
```

```python
# Concatenating the samples with original dataset
dataset_rebal = pd.concat([dataset,new_samples])
print('The new rebalanced dataset is: ')
dataset_rebal
```

```python
new_n = dataset_rebal.shape[0]
new_false,new_true = dataset_rebal['Churn'].value_counts()
new_true_per = np.round((new_true/new_n)*100,decimals=1)
print('The new dataset has {} percent {} Churn instances.'.format(new_true_per,'True'))
```

```python
# finding the mean and standard deviation of training data using Day_mins
day_mins_train = train_data['Day Mins']
mean_train_day_mins = st.mean(day_mins_train)
sd_train_day_mins =st.stdev(day_mins_train)


# finding the mean and standard deviation of test data using Day Mins
day_mins_test = test_data['Day Mins']
mean_test_day_mins = st.mean(day_mins_test)
sd_test_day_mins = st.stdev(day_mins_test)

# Number of isntances in the train and test data
train_n = train_data.shape[0]
test_n = test_data.shape[0]


s1_n1 = (sd_train_day_mins**2)/train_n
s2_n2 = (sd_test_day_mins**2)/test_n



t_Test_values = pd.DataFrame({'Data Type':['Train','Test'],'n':[train_n,test_n],
                             'mean':[mean_train_day_mins,mean_test_day_mins],
                             'Standard Deviation':[sd_train_day_mins,sd_test_day_mins]})

print('The values required for the calculation of t_Test value are: ')
t_Test_values
```

```python
t_Test_val = (mean_train_day_mins - mean_test_day_mins)/np.sqrt(s1_n1+s2_n2)
print('The t_Test value is {}'.format(t_Test_val))
```
