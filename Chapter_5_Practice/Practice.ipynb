```python
# Creating a Training and Test datasets
#Let the test set be 25% of the original dataset
bank_train,bank_test = train_test_split(bank_data,test_size = 0.25,random_state = 7)
```

```python
# Print the data sizes to verify split
n = bank_data.shape[0]
train_n = bank_train.shape[0]
test_n = bank_test.shape[0]
print('The size of original dataset is {}'.format(bank_data.shape))
print('The size of training dataset is {} which is {} percent of original dataset.'
      .format(bank_train.shape,(np.round(train_n/n,decimals=2))*100))
print('The size of test dataset is {} which is {} percent of original dataset.'
      .format(bank_test.shape,(np.round(test_n/n,decimals=2))*100))
```

```python
"""
Validate the partition into training and test sets by graphical and statistical
comparison

The suggested hypothesis test for validating different types of target variables
◦ Numeric – two-sample t-test for the difference in means
◦ Categorical variable with 2 classes – two-sample Z-test for the difference in proportions
◦ Categorical variable with more than 2 classes – test for the homogeneity of proportions
"""

# Two-sample t-Test for difference in means of age attribute
train_mean = np.round(st.mean(bank_train['age']),decimals=2)
test_mean = np.round(st.mean(bank_test['age']),decimals=2)
train_sd = np.round(st.stdev(bank_train['age']),decimals=2)
test_sd = np.round(st.stdev(bank_test['age']),decimals=2)
test_result = [train_mean,test_mean,train_sd,test_sd,train_n,test_n]
print(test_result)
A = (train_sd**2)/train_n
B = (test_sd**2)/test_n
t_data = (train_mean - test_mean)/np.sqrt(A + B)
print('The t-Test value of the splitted data is {}'.format(t_data))
```

```python
# Calculating the p-val
t_val,p_val = stats.ttest_ind(bank_train['age'],bank_test['age'])
t_val
p_val
```
